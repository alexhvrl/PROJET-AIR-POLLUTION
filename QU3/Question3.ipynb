```python
import pandas as pd
import numpy as np # type: ignore
```

```python
data_O3 = pd.read_csv('O3.csv')
O3_MAG = data_O3['Magadino-Cadenazzo'].tolist()
O3_CHA = data_O3['Chaumont'].tolist()
valeurs_O3 = O3_CHA
# Cration d'une liste simule de 8784 valeurs horaires (anne bissextile)
#valeurs = np.random.uniform(low=0, high=100, size=366 * 24)

# Gnrer les timestamps horaires pour 2024 en commenant  01:00
# 366 jours * 24 heures = 8784 heures
# On part de 2024-01-01 01:00
dates = pd.date_range(start='2024-01-01 01:00', periods=len(valeurs_O3), freq='H')

# Crer un DataFrame
df = pd.DataFrame({'Date': dates, 'Valeur': valeurs_O3})

# Dfinir la valeur limite
limite = 120

# Ajouter une colonne pour indiquer si la valeur dpasse la limite
df['Depasse'] = df['Valeur'] > limite

# Grouper par mois et compter les dpassements horaires
resultats = df.groupby(df['Date'].dt.month)['Depasse'].sum()

# Renommer les mois pour affichage (facultatif)
resultats.index = resultats.index.map(lambda x: pd.to_datetime(f'2024-{x:02d}-01').strftime('%B'))

# Afficher le rsultat
#print("Nombre d'heures avec dpassement par mois :")
#print(resultats)

subset_winter_MAG = O3_MAG[0:80*24] + O3_MAG[-24*10:]

av_winter_MAG = np.nanmean(subset_winter_MAG)
print(av_winter_MAG)
av_spring_MAG = np.nanmean(O3_MAG[80*24+1:172*24])
print(av_spring_MAG)
av_summer_MAG = np.nanmean(O3_MAG[172*24+1:266*24])
print(av_summer_MAG)
av_autumn_MAG = np.nanmean(O3_MAG[266*24+1:356*24])
print(av_autumn_MAG)

subset_winter_CHA = O3_CHA[0:80*24] + O3_CHA[-24*10:]

av_winter_CHA = np.nanmean(subset_winter_CHA)
print(av_winter_CHA)
av_spring_CHA = np.nanmean(O3_CHA[80*24+1:172*24])
print(av_spring_CHA)
av_summer_CHA = np.nanmean(O3_CHA[172*24+1:266*24])
print(av_summer_CHA)
av_autumn_CHA = np.nanmean(O3_CHA[266*24+1:356*24])
print(av_autumn_CHA)
```
